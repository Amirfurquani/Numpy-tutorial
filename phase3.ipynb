```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
```

```python

print("some informatin of data")

print(sales_data.ndim)
print(sales_data.shape)
print(sales_data.size)
```

Total sales per year

```python
print(np.sum(sales_data,axis=0))
yearly_sales=np.sum(sales_data[:,1:],axis=0)
print("Yearly sales in zomato:\n",yearly_sales)
```

Minimum sales per restaurent

```python
#min sales per restaurant:
min_sales_per_res=(np.min(sales_data[:,1:],axis=1))
print("min sales per restaurent:\n",min_sales_per_res)
```

```python
#min sales per year:
min_sales_per_year=(np.min(sales_data[:,1:],axis=0))
print("min sales per year:\n",min_sales_per_year)
```

```python
#avg sales per restaurant
avg_sales=(np.average(sales_data[:,1:],axis=1))
print("Average sales per restaurant:\n",avg_sales)
```

```python
#avg sales per year
avg_sales=(np.mean(sales_data[:,1:],axis=0))
print("Average sales per year:\n",avg_sales)
```

Cumulative sum

```python
#for cumulative sum:
cum_sum=np.cumsum(sales_data[:,1:],axis=1)
print("cumulative sum:\n",cum_sum)
```

plot by matlob

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cum_sum,axis=0))
plt.title("Average cumulative sales accross all restaurant:\n")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
```
